```python
# Read the CSV files
import pandas as pd
import numpy as np

df=pd.read_csv('exams.csv')
df

```

```python
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()

df_num_cat= pd.DataFrame()

df_num_cat['gender'] = enc.fit_transform(df['gender'])
df_num_cat['race/ethnicity'] = enc.fit_transform(df['race/ethnicity'])
df_num_cat['parental level of education'] = enc.fit_transform(df['parental level of education'])
df_num_cat['lunch'] = enc.fit_transform(df['lunch'])
df_num_cat['test preparation course'] = enc.fit_transform(df['test preparation course'])
df_num_cat['math score'] = enc.fit_transform(df['math score'])
df_num_cat['reading score'] = enc.fit_transform(df['reading score'])
df_num_cat['writing score'] = enc.fit_transform(df['writing score'])
df_num_cat
```

```python
X=df_num_cat.drop(['writing score'],axis=1)
y=df_num_cat['writing score']
print(X)
print(y)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt_clf = DecisionTreeClassifier(criterion='entropy')
dt_clf.fit(X,y)
```

```python
dt_clf.score(X,y)
y_pred=dt_clf.predict(X)
print(y)
print(y_pred)
```
